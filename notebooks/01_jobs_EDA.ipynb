```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv(
    "../data/cleaned/jobs_clean.csv"
)


df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
top_companies = (
    df["company"]
    .value_counts()
    .head(10)
)

top_companies
```

```python
plt.figure(figsize=(10,5))

top_companies.plot(
    kind="bar"
)

plt.title(
    "Top Hiring Companies in Bengaluru"
)

plt.xticks(
    rotation=45
)

plt.show()
```

```python
from collections import Counter


skills=[]


for s in df["skills"]:

    skills.extend(
        s.split(",")
    )


skill_count = Counter(skills)


skill_df = pd.DataFrame(
    skill_count.items(),
    columns=[
        "skill",
        "count"
    ]
)


skill_df.sort_values(
    "count",
    ascending=False
).head(10)
```

```python
top = skill_df.sort_values(
    "count",
    ascending=False
).head(10)


plt.figure(figsize=(10,5))


plt.bar(
    top["skill"],
    top["count"]
)


plt.xticks(
    rotation=45
)


plt.show()
```
